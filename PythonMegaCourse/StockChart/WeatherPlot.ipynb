```python
import pandas as pd
 
from bokeh.plotting import figure, output_file, show
 
data = pd.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx",sheet_name=0)

data.head()
```

```python
data.describe()
```

```python
# rescale the data
data["Temperature"] = data["Temperature"]/10
data["Pressure"] = data["Pressure"]/10
data.describe()
```

```python
# use bok for bokeh, vs plt for matplotlib

bok = figure(plot_width = 500, plot_height = 400, tools = 'pan')

bok.title.text = "Temperature and Pressure"
bok.title.text_color = "Gray"
bok.title.text_font = "arial"
bok.title.text_font_style = "bold"
bok.title.text_font_size = "16pt"
# bok.xaxis.minor_tick_line_color = None
# bok.yaxis.minor_tick_line_color = None
bok.xaxis.axis_label = "Temperature (C)"
bok.yaxis.axis_label = "Presure (hPa)"

bok.circle(data["Temperature"], data["Pressure"], size = 0.5)
# output_file("Weather.html")
show(bok)
```
